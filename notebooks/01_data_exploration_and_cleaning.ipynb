#### Determinazione possibili colonne da mantenere
Separazione delle fonti pulite e quelle fossili:
Fonti Pulite:
- Wind and Solar
- Hydro, Bioenergy and Other Renewables
- Nuclear
Fonti Fossili:
- Coal
- Gas and Other Fossil

Colonne aggiuntive considerate:
- Electricity demand
- Electricity imports
- Total Generation

```python
ELECTRICITY_GENERATION_VARS = [
    "Clean",
    "Wind and Solar",
    "Hydro, Bioenergy and Other Renewables",
    "Nuclear",
    "Fossil",
    "Coal",
    "Gas and Other Fossil",
    "Total Generation"
]
```

```python
filtered_european_yearly_energy = european_yearly_energy[
    (
        (
            (european_yearly_energy["Category"] == "Electricity generation") &
            (european_yearly_energy["Variable"].isin(ELECTRICITY_GENERATION_VARS))
        )
         |
        (
            (european_yearly_energy["Category"] == "Electricity demand") &
            (european_yearly_energy["Subcategory"] == "Demand")
        )
         |
        (
            european_yearly_energy["Category"] == "Electricity imports"
        )
    )
     &
    (european_yearly_energy["Unit"] == "TWh")
     &
    (european_yearly_energy["Area type"] == "Country or economy")
].copy()

filtered_european_yearly_energy = filtered_european_yearly_energy.dropna(subset=["ISO 3 code", "Value"])
```

### Modifica del Pivot

#### Verifica che esista una sola riga per ogni "Area", "ISO 3 code", "Year", "Variable"

```python
dup_check = (
    filtered_european_yearly_energy
    .groupby(["Area", "ISO 3 code", "Year", "Variable"])
    .size()
    .reset_index(name="count")
)

dup_check[dup_check["count"] > 1].head()
```

#### Modifica del pivot utilizzando come index "Area", "ISO 3 code", "Year"

```python
ordered_columns = [
    # Identificativi
    "Area",
    "ISO 3 code",
    "Year",

    # Domanda / Import
    "Demand",
    "Net Imports",

    # Generazione totale
    "Total Generation",

    # Clean
    "Clean",
    "Nuclear",
    "Wind and Solar",
    "Hydro, Bioenergy and Other Renewables",

    # Fossili
    "Fossil",
    "Coal",
    "Gas and Other Fossil"
]
```

```python
filtered_european_yearly_energy = filtered_european_yearly_energy.pivot_table(
    index=["Area", "ISO 3 code", "Year"],
    columns="Variable",
    values="Value"
).reset_index().reindex(columns=ordered_columns)
filtered_european_yearly_energy.columns.name = None
filtered_european_yearly_energy.head()
```

##### Analisi valori mancanti:

```python
filtered_european_yearly_energy.isna().sum()
```

```python
filtered_european_yearly_energy["Coal"] = (
    filtered_european_yearly_energy["Coal"].fillna(0)
)
```

#### Controllo bilanciamenti (Clean + Fossil == Total Generation, Total Generation + Net Imports == Demand)

```python
import numpy as np

mask = (
    filtered_european_yearly_energy["Clean"].notna() &
    filtered_european_yearly_energy["Fossil"].notna() &
    filtered_european_yearly_energy["Total Generation"].notna()
)

assert np.allclose(
    filtered_european_yearly_energy.loc[mask, "Clean"] + filtered_european_yearly_energy.loc[mask, "Fossil"],
    filtered_european_yearly_energy.loc[mask, "Total Generation"],
    rtol=1e-3,
    atol=1e-2
), "Confronto tra produzione energetica mediante fonti pulite e fossili fallito (Clean + Fossil != Total Generation)"
```

```python
balance_mask = (
    filtered_european_yearly_energy["Total Generation"].notna() &
    filtered_european_yearly_energy["Net Imports"].notna() &
    filtered_european_yearly_energy["Demand"].notna()
)

balance_error = (
    filtered_european_yearly_energy["Total Generation"]
    + filtered_european_yearly_energy["Net Imports"]
    - filtered_european_yearly_energy["Demand"]
)

balance_error.describe()
```

E' ragionevole pensare che gli errori di bilanciamento tra produzione, richiesta e import siano nulli
